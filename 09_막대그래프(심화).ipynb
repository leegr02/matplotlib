# 09. 막대그래프(심화)
구글 검색: matplotlib set_hatch

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #한글폰트설정
matplotlib.rcParams['font.size'] = 15 #폰트 크기
matplotlib.rcParams['axes.unicode_minus']=False #한글 폰트 사용 시 마이너스 글자가 께지는 현상 해결
```

```python
labels = ['강백호','서태웅','정대만'] #이름
values= [190, 187, 184] #키
plt.barh(labels, values)
plt.xlim(175,195)
```

```python
bar=plt.bar(labels, values)
bar[0].set_hatch('/') #////
bar[1].set_hatch('X') # XXXX
bar[2].set_hatch('.') #....
```

```python
bar = plt.bar(labels, values)
plt.ylim(175,195)
for idx, rect in enumerate(bar):
    plt.text(idx,rect.get_height()+0.5, values[idx],ha='center', color='blue')
```
