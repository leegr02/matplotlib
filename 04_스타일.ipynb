# 04. 스타일
구글 검색: matplotlib marker, linestyle, color, markerfacecolor mfc, color picker

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #한글폰트설정
matplotlib.rcParams['font.size'] = 15 #폰트 크기
matplotlib.rcParams['axes.unicode_minus']=False #한글 폰트 사용 시 마이너스 글자가 께지는 현상 해결
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, linewidth=5)
```

## 마커(marker)

```python
plt.plot(x, y, marker='o')
```

```python
plt.plot(x,y, marker='o', linestyle='None')
```

```python
plt.plot(x, y, marker='v', markersize=10)
```

```python
plt.plot(x, y, marker='X', markersize=10)
```

```python
plt.plot(x, y, marker='o', markersize=20, markeredgecolor='red')
```

```python
plt.plot(x, y, marker='o', markersize=20, markeredgecolor='red', markerfacecolor='yellow')
```

## 선 스타일(linestyle)

```python
plt.plot(x, y, linestyle=':')
```

```python
plt.plot(x, y, linestyle='--')
```

```python
plt.plot(x, y, linestyle='-.')
```

```python
plt.plot(x, y, linestyle='-')
```

## 색깔(color)

```python
plt.plot(x, y, color='pink')
```

```python
plt.plot(x, y, color='#ff0000')
```

```python
plt.plot(x, y, color='b')
```

```python
plt.plot(x, y, color='g')
```

## 포맷

```python
plt.plot(x, y, 'ro--') # color, marker, linestyle
```

```python
plt.plot(x, y, 'bv:')
```

```python
plt.plot(x, y, 'go')
```

```python
plt.plot(x, y, color='g', marker='o', linestyle='None')
```

## 축약어

```python
plt.plot(x, y, marker='o', mfc='red', ms='10', mec='blue', ls=':')
```

## 투명도

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, alpha=0.2) # alpha: 투명도(0~1)
```

## 그래프 크기

```python
plt.figure(figsize=(10,5))
plt.plot(x, y)
```

```python
plt.figure(figsize=(5,10))
plt.plot(x, y)
```

```python
plt.figure(figsize=(10,5),dpi=200) #dpi(dot per inch): 확대
plt.plot(x, y)
```

## 배경색

```python
plt.figure(facecolor='yellow')
plt.plot(x, y)
```

```python
plt.figure(facecolor='#a1c3ff')
plt.plot(x, y)
```
